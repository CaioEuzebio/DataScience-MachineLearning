```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(12,10))
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.set_context()
sns.lmplot(x='total_bill', y='tip', data=tips, palette='inferno')
```
